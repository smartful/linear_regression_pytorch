# Régression linéaire

```python
%matplotlib inline
import numpy as np
import torch
import matplotlib.pyplot as plt
```

On simule la création de données immobilières

```python
size = 50
square_meters = np.random.randint(10, 100, size=(size))
square_meters = square_meters.reshape(size, 1)
square_meters = square_meters.astype('float32') 
```

```python
price = (square_meters * 11000) + (np.random.randn(50,1) * 150000)
price = price.astype('float32')
```

```python
plt.figure(figsize=(12,10))
plt.scatter(square_meters, price)
plt.show()
```

```python
square_meters = torch.from_numpy(square_meters)
price = torch.from_numpy(price)
print(square_meters.shape)
print(price.shape)
```

```python
import torch.nn as nn
model = nn.Linear(1, 1)
```

```python
nb_iter = 100
lr = 1e-4
optimizer = torch.optim.SGD(model.parameters(), lr)
```

```python
loss_fn = nn.MSELoss()
```

```python
for i in range(nb_iter):
    price_pred = model(square_meters)
    loss = loss_fn(price_pred, price)
    if (i % 10 == 0):
        print("epoch {} : loss : {}".format(i, loss.item()))
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
```

```python
price_pred = model(square_meters)
print(price[:5])
print(price_pred[:5])
```

```python
with torch.no_grad():
    plt.figure(figsize=(12,10))
    plt.scatter(square_meters.numpy(), price.numpy())
    plt.scatter(square_meters.numpy(), price_pred.numpy(), c='red')
    plt.show()
```

```python
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name, param.data)
```
